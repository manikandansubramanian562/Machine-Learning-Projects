```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
```

```python
df = pd.read_csv("cost_revenue.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
X = DataFrame(df,columns=["production_budget_usd"])
Y = DataFrame(df,columns=["worldwide_gross_usd"])
```

```python
X.head()
```

# Visualization - Feature and Target

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y,alpha=0.3)
plt.title("Film Cost vs Global Revenue")
plt.xlabel("production_budget_usd")
plt.ylabel("worldwide_gross_usd")
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X,Y)
```

# Slope coefficient & Intercept:

```python
regression.coef_
```

```python
regression.intercept_
```

# Visualization - Regression Line

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y,alpha=0.3)
plt.plot(X,regression.predict(X),color="red",linewidth = 4)
plt.title("Film Cost vs Global Revenue")
plt.xlabel("production_budget_usd")
plt.ylabel("worldwide_gross_usd")
plt.xlim(0,450000000)
plt.ylim(0,3000000000)
plt.show()
```

# Goodness of Fit : R-squared

```python
regression.score(X,Y)
```
